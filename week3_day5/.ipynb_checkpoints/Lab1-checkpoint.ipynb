```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('customer_churn.csv')
data
```

```python
data = data[['tenure', 'SeniorCitizen', 'MonthlyCharges', 'Churn']]
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
for x in data.columns:
    print(f'"{x}" ({data[x].dtype}) has {data[x].nunique()} uniqe values:')
    print(data[x].unique())
    print()
```

```python
from sklearn.preprocessing import LabelEncoder  

le = LabelEncoder()
data['Churn'] = le.fit_transform(data['Churn'])
data['Churn'] = data['Churn'].astype('int')
data
```

```python
data['Churn'].astype
```

```python
sns.displot(data['tenure'], kde=True)
plt.show()

sns.displot(data['MonthlyCharges'], kde=True)
plt.show()
```

```python
y = data['Churn']   
X = data.drop('Churn', axis=1)
```

```python
corr_matrix=X.corr() 
fig, ax = plt.subplots(figsize=(5, 4))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(X_train)
X_std_train = transformer.transform(X_train)
X_std_train = pd.DataFrame(X_std_train,columns=X.columns)

X_std_test = transformer.transform(X_test)
X_std_test = pd.DataFrame(X_std_test,columns=X.columns)
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_std_train,y_train)
print(f'Score: {classification.score(X_std_test, y_test)}')
```

Even a simple model will give us more than 70% accuracy. Why?

```python
count_churn = data['Churn'].value_counts()
print(count_churn)
```

For binary classification problems where one class is much more numerous than the other, high accuracy (over 70%) can be achieved even by a simple model.

Simple strategy that predicts the dominant class (e.g., always predicting 0) could achieve high accuracy because most cases belong to this dominant class.

A model may achieve high accuracy, but this does not necessarily mean that it is effective in predicting real-world cases.

The model should also be evaluated using other metrics to get a more complete picture of its performance.

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

predictions = classification.predict(X_std_test)

print(f'Precision: {precision_score(y_test,predictions)}')
print(f'Recall: {recall_score(y_test,predictions)}')
print(f'F1: {f1_score(y_test,predictions)}')
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42,k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_std_train,y_train)
```

```python
class_smote = LogisticRegression(random_state=0,solver='lbfgs').fit(X_train_SMOTE,y_train_SMOTE)
print(f'Score: {class_smote.score(X_std_test, y_test)}')
```

```python
pred_smote = class_smote.predict(X_std_test)

print(f'Precision: {precision_score(y_test,pred_smote)}')
print(f'Recall: {recall_score(y_test,pred_smote)}')
print(f'F1: {f1_score(y_test,pred_smote)}')
```
